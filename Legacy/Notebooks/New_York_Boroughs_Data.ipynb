```python
import pandas as pd 
import numpy as np

```

```python
NY_Boroughs_df = pd.read_csv("New York City Counties.csv")
NY_Boroughs_df.head()
```

```python
NY_Boroughs_df_2 = NY_Boroughs_df.copy()
NY_Boroughs_df_2.head(6)
```

```python
NY_Boroughs_df_2["Population"] = NY_Boroughs_df_2["Population"].apply(lambda x: f"{int(x):,}")
NY_Boroughs_df_2.head(6)
```

```python
NY_Boroughs_df_2.head(6)
```

```python
NY_Boroughs_df_2["County"] = NY_Boroughs_df_2["County"].replace({
    "New York" : "New York City", 
    ".Bronx County, New York" : "Bronx County",
    ".Kings County, New York" : "Kings County (Brooklyn)",
    ".New York County, New York" : "New York County (Manhattan)",
    ".Queens County, New York" : "Queens County",
    ".Richmond County, New York" : "Richmond County (Staten Island)"
})

NY_Boroughs_df_2.head(6)
```

```python
NY_Boroughs_df_2 = NY_Boroughs_df_2.rename(columns={
    "County" : "Borough",
    "Population" : "Population (July, 2024)"
})

NY_Boroughs_df_2.head(6)
```

```python
#Save to a new csv file
NY_Boroughs_df_2.to_csv("NYC Boroughs Data.csv", index=False)
```
